## Some functions to help us out

```python
def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v

def linechart(df, x_dim, y_dim, xaxis, yaxis, title):
    
    fig, ax = plt.subplots(figsize=(15, 7))
    if x_dim == 'index':
        df[y_dim].plot(ax=ax, grid=True)
    else:
        df.set_index(x_dim)[y_dim].plot(ax=ax, grid=True)

    #adds a title and axes labels
    ax.set_title(title)
    ax.set_xlabel(xaxis)
    ax.set_ylabel(yaxis)
    
    #removing top and right borders
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    plt.show()
    
def get_indexed_value(df,row,col):
    msa = row['NAME']
    msa_df = df.loc[df['NAME'] == msa]
    start_val = msa_df.loc[msa_df['year'] == msa_df['year'].min()][col].values[0]
    return (row[col]/start_val)*100

def print_select_linechart(msa_name,var_list):
    single_msa = hhsize_data.loc[hhsize_data['NAME'] == msa_name].sort_values('year',ascending=True)
    msa_06_16 = single_msa.loc[single_msa['year']< 2017]
    msa17_19 = single_msa.loc[single_msa['year']> 2016]
    
    linechart(single_msa, 'year', var_list, 'year', 'total households', 'Household growth in '+msa_name)
    
def print_select_linechart_corr(msa_name,var_list):
    single_msa = hhsize_data.loc[hhsize_data['NAME'] == msa_name].sort_values('year',ascending=True)
    msa_06_16 = single_msa.loc[single_msa['year']< 2017]
    msa17_19 = single_msa.loc[single_msa['year']> 2016]
    
    corr_all_yr = single_msa[var_list+['car_oriented']].corr()[['total_hh_indexed']].reset_index().rename(columns={0:'category','total_hh_indexed':'corr_all_yrs'})
    corr_06_16 = msa_06_16[var_list+['car_oriented']].corr()[['total_hh_indexed']].reset_index().rename(columns={0:'category','total_hh_indexed':'corr_06_16'})
    corr_17_19 = msa17_19[var_list+['car_oriented']].corr()[['total_hh_indexed']].reset_index().rename(columns={0:'category','total_hh_indexed':'corr_17_19'})
    
    corr_master = corr_all_yr.merge(corr_06_16,on='category',how='left')
    corr_master2 = corr_master.merge(corr_17_19,on='category',how='left')
    
    linechart(single_msa, 'year', var_list, 'year', 'total households', 'Household growth in '+msa_name)
    display(corr_master2)
```

## Vehicle availability by household size

```python
myrange = list(range(1,31))

var_list = []
for num in myrange:
    str_num = '{:03d}'.format(num)
    veh_var = 'B08201_'+str_num+'E'
    var_list.append(veh_var)

var_str = ','.join(var_list)

year_list = list(range(2006,2020))
df_list = []

for year in year_list:
    data_url = 'https://api.census.gov/data/'+str(year)+'/acs/acs1?get=NAME,'+var_str+'&for=place:*&in=state:*&key='+MYKEY
    df = pd.read_json(data_url)
    new_header = df.iloc[0] #grab the first row for the header
    df = df[1:] #take the data less the header row
    df.columns = new_header #set the header row as the df header
    df['year'] = year
    display(df.head())
    df_list.append(df)
    
hhsize_data_full = pd.concat(df_list, axis=0, ignore_index=True)
```

```python
hhsize_data = hhsize_data_full
hhsize_data = hhsize_data.dropna(how='any')
hhsize_data[var_list] = hhsize_data[var_list].apply(pd.to_numeric, errors='coerce')

rename_cols = {'B08201_001E':'total_hh','B08201_002E':'total_hh_carfree','B08201_003E':'total_hh_one',
              'B08201_007E':'single_hh','B08201_008E':'single_hh_carfree','B08201_009E':'single_hh_one'}

hhsize_data['total_hh_2more'] = hhsize_data['B08201_004E'] + hhsize_data['B08201_005E'] + hhsize_data['B08201_006E']
hhsize_data['single_hh_2more'] = hhsize_data['B08201_010E'] + hhsize_data['B08201_011E'] + hhsize_data['B08201_012E']
hhsize_data['multi_hh'] = hhsize_data['B08201_013E'] + hhsize_data['B08201_019E'] + hhsize_data['B08201_025E']
hhsize_data['multi_hh_carfree'] = hhsize_data['B08201_014E'] + hhsize_data['B08201_020E'] + hhsize_data['B08201_026E']
hhsize_data['multi_hh_one'] = hhsize_data['B08201_015E'] + hhsize_data['B08201_021E'] + hhsize_data['B08201_027E']
hhsize_data['multi_hh_2more'] = (hhsize_data['B08201_016E'] + hhsize_data['B08201_017E'] + hhsize_data['B08201_018E']) + \
                                (hhsize_data['B08201_022E'] + hhsize_data['B08201_023E'] + hhsize_data['B08201_024E']) + \
                                (hhsize_data['B08201_028E'] + hhsize_data['B08201_029E'] + hhsize_data['B08201_030E'])

hhsize_data.rename(columns=rename_cols, inplace=True)

keep_cols = ['NAME','state','place','year',
             'total_hh','total_hh_carfree','total_hh_one','total_hh_2more',
             'single_hh','single_hh_carfree','single_hh_one','single_hh_2more',
            'multi_hh','multi_hh_carfree','multi_hh_one','multi_hh_2more']

hhsize_data = hhsize_data[keep_cols]
print(hhsize_data.dtypes)

hhsize_data['car_free'] = hhsize_data['total_hh_carfree']
hhsize_data['car_one_families'] = hhsize_data['multi_hh_one']#this is car-lite
hhsize_data['car_one_singles'] = hhsize_data['single_hh_one']
hhsize_data['car_two_plus'] = hhsize_data['total_hh_2more']
hhsize_data['car_oriented'] = hhsize_data['car_one_singles'] + hhsize_data['car_two_plus']

hhsize_data['total_hh_indexed'] = hhsize_data.apply(lambda x: get_indexed_value(hhsize_data,x,'total_hh'), axis=1)
hhsize_data['car_free_i'] = hhsize_data.apply(lambda x: get_indexed_value(hhsize_data,x,'car_free'), axis=1)
hhsize_data['car_one_families_i'] = hhsize_data.apply(lambda x: get_indexed_value(hhsize_data,x,'car_one_families'), axis=1)
hhsize_data['car_one_singles_i'] = hhsize_data.apply(lambda x: get_indexed_value(hhsize_data,x,'car_one_singles'), axis=1)
hhsize_data['car_two_plus_i'] = hhsize_data.apply(lambda x: get_indexed_value(hhsize_data,x,'car_two_plus'), axis=1)
hhsize_data['car_oriented_i'] = hhsize_data.apply(lambda x: get_indexed_value(hhsize_data,x,'car_oriented'), axis=1)

hhsize_min = hhsize_data.sort_values(['NAME','year'])[['NAME','state','place','year','total_hh','total_hh_indexed',
                                                       'car_free_i','car_one_families_i','car_one_singles_i','car_two_plus_i']]
```

## 2019 trends

Peterson looks at these data points on the national and state levels:
- Raw increases/decreases in household numbers
- Increases and decreases in household types - single vs. family
- State vs. national stats
- household vehicle categories as a percentage of all households, overtime
- household vehicle category raw number declines/increases between 2018 and 2019
- household vehicle categories by worker status (I'm not sure how valuable this is... presumably many are?)
- household vehicle categories by age brackets
- MOE (might just need to acknowledge that there are MOE issues likely in the data and that policy makers should be aware before making decisions based on high MOE data)
- Percentage point shifts in car-oriented and car-free lifestyles
- Great Recession trends (2007 - 2010)

"Car-oriented lifestyles in Texas shifted up 0.5 percentage points, while in Washington, they shifted down 0.2 percentage points. For state and urban policymakers who have focused on diversifying Washingtons transportation lifestyles, this indicates progress."

"Pandemics and recessions also mess with household formation, relevant because the form that households take affect their transportation lifestyles and ability to adapt to changing circumstances."

We will want to mention that the pandemic may push more intergenerational living situations, children living with parents and parents with children, thus changing many people's access to vehicles. Various situations are feasible: young adult with no car moving from city to suburbs with parents who do have a car, elderly parent with no car moving in with married son or daughter who have multiple cars, previously single adult w/o car moves in with vehicle-owning roommate or partner due to financial burdens. The point is that the trends we see in the next few years will be very interesting and certainly complicated. 

"If the pattern of the Great Recession repeats, there will soon be many more families with only one vehicle."

There are lots of financial hardship surveys out there to pull from: Bankrate's [Jan. 2020 Financial Security Index](https://www.bankrate.com/banking/savings/financial-security-january-2020/) found that 59% of US adults wouldn't be able to cover a \\$1,000 unexpected expense. [Federal Reserve](https://www.federalreserve.gov/publications/2019-economic-well-being-of-us-households-in-2018-dealing-with-unexpected-expenses.htm) has a similar report on \\$400 unexpected expenses and Marketplace did an Edison poll that found 60% can't deal with \\$1,000 unexpecedted expense and more than 40\\% would struggle with \\$250 unexpected.

But Peterson mentions that there is a brightside: "For vulnerable households, many of the pandemics transportation trends support car-free and car-lite living. The spread of home delivery of goods including groceries, work from home for office workers, and telehealth and other online services may make shedding vehicles easier."

```python
#possible values of veh_class: 
#Vars used in story: car_free, car_one_families, car_one_singles, car_two_plus, car_oriented (combo of car_one_singles + car_two_plus)
#Households with only one person: single_hh_carfree, single_hh_one, single_hh_2more
#Households with 2+ persons: multi_hh_carfree, multi_hh_one, multi_hh_2more
def get_veh_class_household_diff(veh_class):
    if veh_class == '' or veh_class == None:
        print('Here are your options:')
        print('Vars used in story: car_free, car_one_families, car_one_singles, car_two_plus, car_oriented (combo of car_one_singles + car_two_plus)')
        print('Households with only one person: single_hh_carfree, single_hh_one, single_hh_2more')
        print('Households with 2+ persons: multi_hh_carfree, multi_hh_one, multi_hh_2more')
    else:
        household_growth = pd.pivot_table(hhsize_data, values=veh_class, index=['NAME'],
                        columns=['year'], aggfunc=np.sum)
        return household_growth.diff(axis=1)
```

Let's create a function that will take a city name and return a profile of all the data bits we've compiled so far.

```python
def get_city_profile(city):
    cols = ['indicator']+list(range(2006,2020))
    overtime_df = pd.DataFrame(columns=cols)
    
    #Total households
    hh_growth = household_growth.loc[household_growth.index == city]
    hh_growth['indicator'] = 'Total households'
    overtime_df = overtime_df.append(hh_growth, ignore_index=True)
    
    #Raw increases/decreases in household numbers
    hh_chg = household_diff.loc[household_diff.index == city]
    hh_chg['indicator'] = 'Change in total households'
    overtime_df = overtime_df.append(hh_chg, ignore_index=True)
    
    #Increases and decreases in household types - single vs. family
    multi_hh_diff_city = multi_hh_diff.loc[multi_hh_diff.index == city]
    multi_hh_diff_city['indicator'] = 'Change in multi-person households'
    overtime_df = overtime_df.append(multi_hh_diff_city, ignore_index=True)
    
    single_hh_diff_city = single_hh_diff.loc[single_hh_diff.index == city]
    single_hh_diff_city['indicator'] = 'Change in single-person households'
    overtime_df = overtime_df.append(single_hh_diff_city, ignore_index=True)
    
    print('Household data about '+ city)
    display(overtime_df)
    
    
    #household vehicle categories as a percentage of all households, overtime
    #car_free, car_one_families, car_one_singles, car_two_plus, car_oriented
    print('Vehicle categories as a percentage of all households')
    display(hh_growth.set_index('indicator').pct_change(axis='columns'))
    veh_class_perc_city = veh_class_perc.loc[veh_class_perc.index.get_level_values('NAME') == city]
    
    #pct_change from previous year in vehicle categories
    print('Percent change from previous year in vehicle categories')
    print('Average total household growth 2006-2019:',hh_growth.set_index('indicator').pct_change(axis='columns').mean(axis=1, numeric_only=True)['Total households'])
    veh_class_growth_pctchg_city = veh_class_growth.loc[veh_class_growth.index.get_level_values('NAME') == city].pct_change(axis='columns')
    veh_class_growth_pctchg_city['avg_06_19'] = veh_class_growth_pctchg_city.mean(axis=1)
    display(veh_class_growth_pctchg_city)
    
    #all_vehcat = get_veh_class_household_diff('car_free')
    #all_vehcat_city = all_vehcat.loc[all_vehcat.index == city]
    #all_vehcat_city['indicator'] = 'Chg in car-free households'
    #overtime_df = overtime_df.append(all_vehcat_city, ignore_index=True)
    #
    #all_vehcat = get_veh_class_household_diff('car_one_families')
    #all_vehcat_city = all_vehcat.loc[all_vehcat.index == city]
    #all_vehcat_city['indicator'] = 'Chg in family w/ 1 car (car-lite) households'
    #overtime_df = overtime_df.append(all_vehcat_city, ignore_index=True)
    #
    #all_vehcat = get_veh_class_household_diff('car_two_plus')
    #all_vehcat_city = all_vehcat.loc[all_vehcat.index == city]
    #all_vehcat_city['indicator'] = 'Chg in households w/ 2+ cars'
    #overtime_df = overtime_df.append(all_vehcat_city, ignore_index=True)
    #
    #all_vehcat = get_veh_class_household_diff('car_oriented')
    #all_vehcat_city = all_vehcat.loc[all_vehcat.index == city]
    #all_vehcat_city['indicator'] = 'Chg in car-oriented households'
    #overtime_df = overtime_df.append(all_vehcat_city, ignore_index=True)
    #
    #all_vehcat = get_veh_class_household_diff('car_one_singles')
    #all_vehcat_city = all_vehcat.loc[all_vehcat.index == city]
    #all_vehcat_city['indicator'] = 'Chg in single households w/ 1 car'
    #overtime_df = overtime_df.append(all_vehcat_city, ignore_index=True)
    
    
    ##household vehicle categories by age brackets
    #print('')
    #print('Carfree lifestyles by age brackets')
    ##df.index.get_level_values(1)
    #display(age_class_growth.loc[age_class_growth.index.get_level_values(0) == city])
    
    #Great Recession trends (2007 - 2010)
    print('')
    var_list = ['total_hh_indexed','car_free_i','car_one_families_i','car_one_singles_i','car_two_plus_i']
    print_select_linechart_corr(city,var_list)
    
    print('')
    var_list = ['total_hh','car_free','car_one_families','car_one_singles','car_two_plus']
    print_select_linechart(city,var_list)
```
